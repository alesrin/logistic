### EDA ( Análisis exloratorio de los datos ): 

Técnica utilizada en estadística y ciencia de datos para analizar y resumir conjuntos de datos, principalmente mediante técnicas visuales. Es el paso intermedio entre la carga de datos y el preprocesamiento. Hay varias visualizaciones que se pueden utilizar en el análisis exploratorio de datos:

- `Histograma`: Un histograma es una representación gráfica de la distribución de un conjunto de datos numéricos. Muestra la frecuencia de los datos en intervalos, lo que permite analizar la forma de la distribución, la centralidad y la dispersión.

- `Diagrama de caja (box plot)`: Un diagrama de caja es una representación gráfica que muestra la mediana, los cuartiles y los valores atípicos de un conjunto de datos numéricos. Es útil para visualizar la dispersión y simetría de los datos, así como identificar posibles valores atípicos.

- `Diagrama de dispersión (scatter plot)`: Un diagrama de dispersión es una representación gráfica de dos variables numéricas, en la que cada punto representa una observación y sus coordenadas corresponden a los valores de ambas variables. Es útil para visualizar la relación entre dos variables y detectar posibles correlaciones, patrones o tendencias.

- `Gráfico de barras`: Un gráfico de barras es una representación gráfica de datos categóricos en la que cada categoría se representa mediante una barra, y la altura o longitud de la barra corresponde a la frecuencia, proporción o medida de interés de esa categoría. Es útil para comparar diferentes categorías y analizar datos categóricos.

- `Gráfico de líneas`: Un gráfico de líneas es una representación gráfica de datos numéricos a lo largo del tiempo, en el que los puntos se conectan mediante líneas. Es útil para analizar tendencias, patrones y fluctuaciones a lo largo del tiempo, así como comparar series temporales de diferentes variables.

- `Gráfico de pastel (pie chart)`: Un gráfico de pastel es una representación gráfica de datos categóricos en la que cada categoría se representa mediante un sector circular, y el tamaño del sector es proporcional a la frecuencia, proporción o medida de interés de esa categoría. Es útil para visualizar la composición de un conjunto de datos y comparar proporciones de diferentes categorías.

- `Gráfico de densidad (density plot)`: Un gráfico de densidad es una representación suavizada de la distribución de un conjunto de datos numéricos, que estima la función de densidad de probabilidad subyacente. Es útil para analizar la forma de la distribución y comparar distribuciones de diferentes conjuntos de datos.

| Tipo de gráfico       | Descripción                                                                                  | Ejemplos de gráficos disponibles en Matplotlib                                      |
|-----------------------|----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------|
| **Gráficos univariantes** | Representan una sola variable, analizando su distribución, centralidad y dispersión.        | Histogramas, Gráficos de barras, Gráficos de pastel, Gráficos de densidad         |
| **Gráficos bivariantes** | Representan dos variables para analizar la relación entre ellas, identificando patrones.      | Diagramas de dispersión (`scatter`), Gráficos de líneas, Gráficos de barras apiladas |
| **Gráficos multivariantes** | Representan tres o más variables, analizando relaciones complejas y tendencias entre ellas.  | Gráficos de burbujas, Gráficos de barras apiladas, Gráficos de coordenadas paralelas |



### Gráficos de líneas

Representación gráfica de datos numéricos. `plot()`

````python
import matplotlib.pyplot as plt


# Definimos un conjunto de datos
x = [1,2,3,4,5] # Valores de x
y = [1,4,9,16,25] # Valores de y

# Creamos la gráfica de líneas
plt.plot(x, y, label="Valores", color='red', linestyle="dotted") 

# Añadimos un título y etiquetas
plt.title("Gráfica de líneas")
plt.xlabel("x")
plt.ylabel("y")

# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
````

### Grafícos de dispersión

Nos permiten visualizar la relación entre dos variables. `scatter()`

````python
# Creamos el gráfico de dispersión
plt.scatter(x, y, label="Puntos de dispersión", color='blue', marker='o')

# Añadimos un título y etiquetas
plt.title("Gráfico de dispersión")
plt.xlabel("x")
plt.ylabel("y")

# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
````

### Gráficos de barras
Nos permiten comparar categorías o clases. `bar()`

````python
# Definimos nuevos datos
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [20, 35, 30, 35, 27]

# Creamos un gráfico de barras
plt.bar(categorias, valores, color='green')

# Añadimos un título y etiquetas
plt.title("Gráfico de barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")

# Mostrar la gráfica
plt.show()
````

### Histogramas 

Representación gráfica de la distribución de datos numéricos. `hist`

````python
# Definimos nuevos datos
datos = [ 1, 2, 2, 3, 4, 4, 4, 5, 6, 6, 7, 8, 8, 9, 10]

# Creamos un histograma
plt.hist(datos, bins=10, color='orange', edgecolor='black')
# datos: Datos a representar
# bins: Número de divisiones, barras, contenedores con los que se representará la frecuencia en el histograma

# Añadimos un título y etiquetas
plt.title("Histograma")
plt.xlabel("Datos")
plt.ylabel("Frecuencia")

# Mostrar la gráfica
plt.show()
````

### GRÁFICO PASTEL

Representamos datos categóricos en los que cada sector va a representar una categoría dentro de mi conjunto de datos. `pie()`

````python
# Definimos nuevos datos
tamanos = [10, 20, 30, 40, 50]
valores = [1, 2, 3, 4, 5]

# Creamos un gráfico circular
plt.pie(valores, labels=tamanos, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green', 'yellow', 'purple'])
#tamanos: Las proporciones de cada sección
#valores: Etiquetas de cada sección
#autopct: Formato para mostrar los porcentajes
#startangle: Ángulo inicial de la gráfica

plt.title("Gráfico circular")

# Mostrar la gráfica
plt.show()
````

### Gráfico de densidad

Nos muestran la distribución de un conjunto de datos NUMÉRICOS. `kdeplot()`

````python
import seaborn as sns
import numpy as np

# Generar un conjunto de datos
datos = np.random.normal(loc=0, scale=2, size=1000)

# Creamos un gráfico de densidad
sns.kdeplot(datos, color='red')

# Añadimos un título y etiquetas
plt.title("Gráfico de densidad")
plt.xlabel("Datos")
plt.ylabel("Densidad")

# Mostrar la gráfica
plt.show()

````
